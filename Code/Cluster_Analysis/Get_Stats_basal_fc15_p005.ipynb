# P_Value cutoff (<0.05) has been applied for DEGs

```python
import pandas as pd
import os
```

```python
#directories 
results_dir_5ht = "../../Results/Results_no_ptprc_adgre1/Cluster_Analysis_avg_link_gene_cell_corr/DEG_p005/DEG_TargetGenes_Sig_Regulons_fc15_p005/5Ht"
results_dir_6ho = "../../Results/Results_no_ptprc_adgre1/Cluster_Analysis_avg_link_gene_cell_corr/DEG_p005/DEG_TargetGenes_Sig_Regulons_fc15_p005/6Ho"
output_dir = "../../Results/Results_no_ptprc_adgre1/Cluster_Analysis_avg_link_gene_cell_corr/DEG_p005/DEG_TargetGenes_Sig_Regulons_fc15_p005"
```

```python
#Function to remove duplicates from target gene list
def fix_duplicates(gene_list):
    if len(gene_list) == 0:
        return gene_list
    if '+' in gene_list[0]:
        gene_name = gene_list[0].split('(')[0]
        if gene_name in gene_list:
            del gene_list[gene_list.index(gene_name)]
    return gene_list
```

# 5Ht

```python
clusters = os.listdir(results_dir_5ht)
if 'Unique_regulon_clust_results' in clusters:
    del clusters[clusters.index('Unique_regulon_clust_results')]
    
for cluster_file in clusters:
    regulons = os.listdir(f"{results_dir_5ht}/{cluster_file}")
    regulon_names = [r.split('.')[0] for r in regulons]
    
    if "S" in cluster_file:
        curr_df = pd.DataFrame(columns=['Total Target Genes','Number of Significant Target Genes (5Ht_6Ho)',
                           'Frequency of Significant Target Genes ((5Ht_6Ho))','Significant Target Genes (5Ht_6Ho)',
                           'Number of Significant Target Genes (SameCell)','Frequency of Significant Target Genes (SameCell)'
                          ,'Significant Target Genes (SameCell)'],index = regulon_names)
        for i,regulon in enumerate(regulons):
            x = pd.read_csv(f"{results_dir_5ht}/{cluster_file}/{regulon}")
            sig_genes_1 =  list(x[x['5Ht_6Ho_DEG']==1]['Gene'])
            sig_genes_1 = fix_duplicates(sig_genes_1)
            
            sig_genes_2 = list(x[x['5Ht_same_celltype_DEG']==1]['Gene'])
            sig_genes_2 = fix_duplicates(sig_genes_2)
            
            data = [f"{len(x)}",len(sig_genes_1),len(sig_genes_1)/len(x),sig_genes_1,len(sig_genes_2),
                    len(sig_genes_2)/len(x),sig_genes_2]
            
            curr_df.loc[regulon_names[i]] = data
        curr_df.index.name = 'Significant_Regulons'
        curr_df = curr_df.sort_values(by='Frequency of Significant Target Genes ((5Ht_6Ho))', ascending=False)
        output_path = f'{output_dir}/Stats/5ht/'
        os.makedirs(output_path, exist_ok=True)
        curr_df.to_csv(f'{output_path}/{cluster_file}_5Ht.csv',index=True,quoting=1)
    
    elif "U1" in cluster_file:
        curr_df = pd.DataFrame(columns=['Total Target Genes','Number of Significant Target Genes (SameCell)',
                           'Frequency of Significant Target Genes ((SameCell))','Significant Target Genes (SameCell)',
                           'Number of Significant Target Genes (Global)','Frequency of Significant Target Genes (Global)'
                          ,'Significant Target Genes (Global)'],index = regulon_names)
        for i,regulon in enumerate(regulons):
            x = pd.read_csv(f"{results_dir_5ht}/{cluster_file}/{regulon}")
            sig_genes_1 =  list(x[x['5Ht_same_celltype_DEG']==1]['Gene'])
            sig_genes_1 = fix_duplicates(sig_genes_1)
            
            sig_genes_2 = list(x[x['5Ht_global_DEG']==1]['Gene'])
            sig_genes_2 = fix_duplicates(sig_genes_2)
            
            data = [f"{len(x)-1}",len(sig_genes_1),len(sig_genes_1)/len(x),sig_genes_1,len(sig_genes_2),
                    len(sig_genes_2)/len(x),sig_genes_2]

            curr_df.loc[regulon_names[i]] = data
        curr_df.index.name = 'Significant_Regulons'
        curr_df = curr_df.sort_values(by='Frequency of Significant Target Genes ((SameCell))', ascending=False)
        output_path = f'{output_dir}/Stats/5ht/'
        os.makedirs(output_path, exist_ok=True)
        curr_df.to_csv(f'{output_path}/{cluster_file}_5Ht.csv',index=True,quoting=1)
    
    elif 'Unique' in cluster_file:
        curr_df = pd.DataFrame(columns=['Total Target Genes','Number of Significant Target Genes (5Ht_6Ho)',
                           'Frequency of Significant Target Genes ((5Ht_6Ho))','Significant Target Genes (5Ht_6Ho)',
                           'Number of Significant Target Genes (SameCell)',
                           'Frequency of Significant Target Genes ((SameCell))','Significant Target Genes (SameCell)',
                           'Number of Significant Target Genes (Global)','Frequency of Significant Target Genes (Global)'
                          ,'Significant Target Genes (Global)'],index = regulon_names)
        
        for i,regulon in enumerate(regulons):
            x = pd.read_csv(f"{results_dir_5ht}/{cluster_file}/{regulon}")
            sig_genes_1 =  list(x[x['Basal_5Ht_6Ho_DEG']==1]['Gene'])
            sig_genes_1 = fix_duplicates(sig_genes_1)
            
            sig_genes_2 =  list(x[x['5Ht_Basal_same_celltype_DEG']==1]['Gene'])
            sig_genes_2 = fix_duplicates(sig_genes_2)
            
            sig_genes_3 = list(x[x['5Ht_Basal_global_DEG']==1]['Gene'])
            sig_genes_3 = fix_duplicates(sig_genes_3)
            
            data = [f"{len(x)}",len(sig_genes_1),len(sig_genes_1)/len(x),sig_genes_1,len(sig_genes_2),
                    len(sig_genes_2)/len(x),sig_genes_2,len(sig_genes_3),len(sig_genes_3)/len(x),sig_genes_3]
            
            
            curr_df.loc[regulon_names[i]] = data
        curr_df.index.name = 'Significant_Regulons'
        curr_df = curr_df.sort_values(by='Frequency of Significant Target Genes ((5Ht_6Ho))', ascending=False)
        output_path = f'{output_dir}/Stats/5ht/'
        os.makedirs(output_path, exist_ok=True)
        curr_df.to_csv(f'{output_path}/{cluster_file}_5Ht.csv',index=True,quoting=1)
        
```
